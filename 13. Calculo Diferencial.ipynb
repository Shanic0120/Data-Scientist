# Cálculo diferencial

El cálculo es realizar operaciones de una manera dada para llegar a un resultado. Cuando se habla de diferencial se habla la rama de cálculo infinidecimal, que estudia la variación de las funciones cuando tiene cambios muy pequeños.



## [Límite](https://www.youtube.com/watch?v=o2UTk8bsLS0)

Un límite es una herramienta matemática que nos indica que pasa si se aproxima los valores de una función a un punto, es interesante ya que en este cáculo se desea medir variaciones muy pequeñas en los valores de una función.

El límite de $f(x)$ cuando $x$ tiende a $a$ es $L$: $$\lim_{x\rightarrow a }f(x)=L$$

Un límite lateral es cuando se evalúa el límite por la "parte" izquierda o derecha de una función. Un límite L solo existe cuando el límite por izquierda es igual o muy similar al límite por la derecha

$$\lim_{x\rightarrow a^- }f(x)=L$$
$$\lim_{x\rightarrow a^+ }f(x)=L$$

Ejemplo:



Tenemos esta función, el punto A está resaltado porque la función no puede tomar el valor $x=2$, debido a que sería indeterminado. Los límites recordemos que es que sucede cuando la función se aproxima a un valor: 

$$\lim_{x\rightarrow 2 }f(x)=\lim_{x\rightarrow 2 }\frac{x^2-4}{x-2}$$

$$\text{Factorizamos por diferencia de cuadrados:}$$

$$\lim_{x\rightarrow 2 }\frac{(x-2)(x+2)}{(x-2)}=\lim_{x\rightarrow 2 }(x+2)=\lim_{x\rightarrow 2 }(2+2)=4$$

## Derivada

Imaginemos una función cualquiera:



Donde $h$ es la distancia horizontal entre A y B, por lo que podría decirse que:

$A=(x,f(x))$

$B=(x+h, f(x+h))$

Recordemos que la pendiente está dada por:

$$m=\frac{y_2-y_1}{x_2-x_1}$$

O sea, para dos puntos A y B:

$$m=\frac{f(x+h)-f(x)}{x+h-x}=\frac{f(x+h)-f(x)}{h}$$

La idea del límite es aproximar lo más posible el punto B hacia el punto A.



Podemos observar que al posicionar B en otro punto de la función, la recta pendiente cambia, por lo que se puede decir que mientras más cercanos estemos al punto se va a obtener la pendiente de la recta tangente. Para ello $h$ debe tender a 0, esto es la derivada:

$$y'=f'(x)=\frac{dy}{dx}=\lim_{h\rightarrow 0 }\frac{f(x+h)-f(x)}{h}$$
$$y'=\lim_{\Delta x\rightarrow 0 }\frac{\Delta y}{\Delta x}=\frac{dy}{dx}$$

### Razón de cambio



Tomemos esta función donde $f(x)$ es el salario de una persona y $x$ son los años de experiencia, la derivada se puede tomar como qué tanto va a cambiar el salario en un pasado o futuro.

$$\frac{dy}{dx}=f'(x)=2x$$

Esta derivada nos dice que la función está creciendo a un ritmo de $2x$

### Implementación de una derivada discreta

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def df(f, x):
    h = 0.000001
    return (f(x+h) - f(x)) / h
```

```python
x = np.linspace(-10, 10, num=1000)
y = f(x)

plt.plot(x, y)
plt.plot(x, df(f, x), "r", alpha=0.6)
plt.grid()
plt.show()
```

### ¿Qué es un máximo y un mínimo?

En matemáticas, un máximo y un mínimo son valores extremos de una función que representan los puntos más altos y más bajos, respectivamente. Estos extremos son importantes en el análisis de funciones, ya que brindan información sobre los puntos críticos y las características de la función. A continuación, se describen con más detalle cada uno de ellos:

**Máximo:**
Un máximo de una función es el valor más alto que la función alcanza en un determinado intervalo o en todo su dominio. Formalmente, sea f una función definida en un conjunto D, se dice que f tiene un máximo en el punto c si f(c) es mayor o igual que f(x) para todo x en D cercano a c. Un máximo absoluto es aquel que es el valor más alto de la función en todo su dominio, mientras que un máximo relativo es aquel que es el valor más alto en un intervalo específico.

**Mínimo:**
Un mínimo de una función es el valor más bajo que la función alcanza en un determinado intervalo o en todo su dominio. Formalmente, sea f una función definida en un conjunto D, se dice que f tiene un mínimo en el punto c si f(c) es menor o igual que f(x) para todo x en D cercano a c. Un mínimo absoluto es aquel que es el valor más bajo de la función en todo su dominio, mientras que un mínimo relativo es aquel que es el valor más bajo en un intervalo específico.

### Optimizando un modelo


**¿Por qué optimizar?**

Es necesario tener en cuenta la optimización buscando una función objetivo, esto significa que, con ayuda de las derivadas y los límites podrémos encontrar el área más óptima para el desarrollo de las "oficinas" para esto se hace lo siguiente:

Paso 01: entender el área y perimetro del espacio:

$A = x * y$

$P = x + y + x = 2x + y$

Paso 02: Tener un área, en este caso de muro, como restricción (50ml de perimetro en este caso)

$$2x + y = 50$$

Paso 03: Buscar la función objetivo, en este caso y (es la función que queremos optimizar)

$$y = 50-2x$$

Paso 04: Reescribir en terminos de la dependiente área en terminos de x

$$A(x) = x(50-2x)$$ 
$$ A(x) = -2x² + 50x$$

Paso 05: optimizar buscando su derivada e igualar a 0 (se iguala porque queremos saber en qué punto la x está en el punto más alto o más bajo)

$$A'(x) = -4x + 50 = 0$$

$$-4x + 50 = 0$$

$$x = \frac{-50}{-4}= \frac{25}{2} \text{ (mínima expresión)}$$

Paso 06: Buscar datos a la izquierda y a la derecha

$$A(x) = -2x² + 50x$$

A la izquierda:

$$A(12.45) = -2(12.45)^2 + 50(12.45) = \frac{1}{5} \text{ (pendiente positiva)}$$

A la derecha:

$$A(12.55) = -2(12.55)^2 + 50(12.55) = -\frac{1}{5} \text{ (pendiente negativa)}$$

Conclusión: 12.50 es igual a 0 como punto más alto (máximo) total

$$A = 12.5 * 25 = 312.5 mt^2 \text{ (área más optimizada)}$$
$$P = 12.5² + 25 = 50$$

### ¿Cómo son las derivadas en las funciones de activación?

Es importante la momento de usar descenso al gradiante, ya que este busca los mínimos para optimizar, por lo que si las funciones son muy complejas.

Las funciones de activación más comunes son:

- Función sigmoide
- Función tangente hiperbólica (tanh)
- Función ReLU (Rectified Linear Unit)
- Función Leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def df(f, x = np.linspace(-10, 10, 1000)):
    h = 0.000001
    return (f(x+h) - f(x)) / h
```

#### Función lineal

$$f(x)=mx+b$$

Derivada:

$$f(x)=m$$

```python
def f(x):
    return x

x = np.linspace(-10, 10, 1000)

plt.plot(x, f(x))
plt.plot(x, df(f), "r", alpha=0.6)
plt.grid()
```

#### Función heaviside

$$H(x)=\begin{cases}
0 & \text{para, } x < 0 \\
1 & \text{para, } x \ge 0
\end{cases}$$

Derivada:

$$f'(x)=0$$

```python
def H(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            y[idx] = 1
    return y

plt.plot(x, H(x))
plt.plot(x, df(H), "r", alpha=0.6)
plt.grid()
```

#### Función sigmoide

$$y=\frac{1}{1+e^{-z}}$$

Derivada:

$$f'(x)=f(x)(1-f(x))$$

```python
def f(x):
    return 1/(1+np.exp(-x))

plt.plot(x, f(x))
plt.plot(x, df(f), "r", alpha=0.6)
plt.grid()
```

#### Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

Derivada:

$$f'(x)=1-f(x)^2$$

```python
def tanh(x):
    return np.tanh(x)

plt.plot(x, tanh(x))
plt.plot(x, df(tanh), "r", alpha=0.6)
plt.grid()
```

#### Función ReLu

$$R(x)=\begin{cases}
0 & \text{para, } x < 0 \\
x & \text{para, } x \ge 0
\end{cases}$$

Derivada:

$$R(x)=\begin{cases}
0 & \text{para, } x < 0 \\
1 & \text{para, } x > 0
\end{cases}$$

```python
def relu(x):
    return np.maximum(x, 0)

plt.plot(x, relu(x))
plt.plot(x, df(relu), "r", alpha=0.6)
plt.grid()
```
