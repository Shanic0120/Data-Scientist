## Sublot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.subplot(1, 2, 1) # rows, columns, index actual
plt.plot(x, y, "r--")
plt.plot(y, x, "b:")
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```

## Orientado a objetos

Es más complejo de usar, pero mucho más personalizable

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # Pos, pos, tamaño, tamaño
axes.plot(x, y)
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
ax2 = fig.add_axes([0.25, 0.7, 0.4, 0.3])
ax1.plot(x, y)
ax2.plot(y, x, "r")
ax2.set_facecolor("green") # Elegir background
ax2.set_ylabel("y")
ax2.set_xlabel("x")
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
ax1.plot(x, y)
ax2.scatter(y, x)

plt.show()
```

### Subplots

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y)
ax[1].plot(y, x)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y)
ax2.plot(y, x)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax[0, 0].plot(x, y, label="$sin(x)$")
ax[0, 0].set_title("Relación de X - Y")
ax[0, 0].set_xlabel("X")
ax[0, 0].set_ylabel("Y")
ax[0, 0].legend(loc = "lower left") # Para organizarlo por posición: bbox_to_anchor=(0.5,0.2)

ax[1, 1].plot(y, x)
plt.show()
```

```python
nrows = 3
ncols = 3
fig, ax = plt.subplots(nrows, ncols)

for row in range(nrows):
    for col in range(ncols):
        ax[row, col].plot(x, y)
```

## [Estilos](https://matplotlib.org/cheatsheets/_images/cheatsheets-1.png)




```python
for i in plt.style.available:
    plt.style.use(i)
    plt.plot(x, y, "bo-", label="$sin(x)$")
    plt.plot(x, y*x, "rs-", label="$sin(x)^2$")
    plt.plot(y, x, "g--", label="$sin^{-1}(x)$")
    plt.title(i)
    plt.legend()
    plt.show()
```
