# Importar librer√≠as

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot
Usado para graficar dos variables numericas, generalmente para series de tiempo, porcentajes, etc.

```python

x = [i for i in range(2016,2022)]

# Pais 1
y1 = [45, 46, 46, 48, 50, 51]

# Pais 2
y2 = [46, 44, 45, 46, 49, 46]

plt.figure(figsize=(10,6)) # Cambiar tama√±o de la figura

plt.plot(x, y1, 
         marker = "o",
         linestyle = "--",
         color = "r",
         label = "Colombia")

plt.plot(x, y2, 
         marker = "d",
         markerfacecolor="r",
         linestyle = "-",
         alpha = 0.8,
         linewidth = 4,
         color = "g",
         label = "Peru")

plt.xlabel("A√±os")
plt.ylabel("Poblaci√≥n (M)")
plt.title("A√±os vs Poblaci√≥n")
plt.legend(loc="lower right")

plt.yticks([45,48,51])

#plt.savefig("nombre_archivo.png")

plt.show()
```

# Bar Plots
Necesitan un eje categorico y otro numerico, usado para mostrar frecuencia de datos

```python
x = ["Argentina", "Colombia", "Peru", "Brasil"]
y = [40, 50, 27, 57]

plt.bar(x, y, width=0.5, color = ["r","b"],)
plt.axhline(y = np.average(y),label='Media', color='red', linestyle='--')
plt.axhline(y = np.median(y), label='Mediana', color='#a3b18a', linestyle='-.')
plt.xticks(np.arange(6), # Numero de posiciones
            ["Argentina", "Colombia", "Peru", "Brasil", "null", "null"],
            rotation=45) # Configurar el eje x
plt.legend()
plt.show()
```

```python
plt.barh(x, y, color = ["r","b"],)
```

# Piecharts
Usado con datos numericos junto a sus labels, usado para representar porcentajes, poblaciones, etc.

```python
plt.pie(y, labels = x)

plt.show()
```

# Histogramas
Necesita los "bins" que representa el tama√±o que va a tener cada barra y los datos numericos, se usa para visualizar la distribuci√≥n de los datos

```python
edades = [15, 16, 17, 20, 21, 21, 22, 23, 24 ,25, 26, 30, 31, 32, 35]
bins = [15, 20, 25, 30, 35]

plt.hist(edades, bins = bins,
         edgecolor="red",
         width=3,
         histtype="bar")
plt.show()
```

```python
plt.hist(edades, bins = 4,
         edgecolor="red",
         histtype="step")
plt.show()
```

```python
plt.hist(edades, bins = bins,
         edgecolor="red",
         histtype="stepfilled")
plt.show()
```

# Boxplots
Usado para detectar elementos atipicos (outlayers)

```python
edades = [15, 16, 17, 20, 21, 21, 22, 23, 24 ,25, 26, 30, 31, 32, 35, 60, 70]

plt.boxplot(edades,
            vert=True, # Verticalidad
            patch_artist=True, # Pintarlo
            notch=True, # Mostrar como se distribuyen (achicarlo)
            showfliers=True # Mostrar datos at√≠picos
            )
plt.show()
```

# Scatterplot
Relaci√≥n entre dos variables numericas

```python
a = [1, 2, 3, 4, 5, 4, 3, 2, 4, 5, 6, 5, 7]
b = [7, 2, 2, 5, 5, 7, 3, 2, 1, 4, 6, 3, 2]

plt.scatter(a, b)
plt.show()
```

# Subplots
Visualizar multiples graficos en la misma vista

```python
x = [i for i in range(2016,2022)]

y1 = [45, 46, 46, 48, 50, 51]
y2 = [46, 44, 45, 46, 49, 46]

fig, ax = plt.subplots(1, 2,
                       sharey = True)

ax[0].plot(x, y1, color = "g", label = "Colombia")
ax[1].plot(x, y2, color = "r", label = "Peru")

ax[0].legend()
ax[1].legend()

plt.show()
```

# GridSpec

```python
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure()
gs = GridSpec(4, 4)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

N = 50 
x = np.random.rand(N)
y = np.random.rand(N)

ax_scatter.scatter(x,y)
ax_hist_x.hist(x, orientation = 'horizontal')
ax_hist_y.hist(y)

plt.show()
```
