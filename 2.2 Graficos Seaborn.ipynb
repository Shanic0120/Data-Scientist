# [Seaborn](https://deeplearning.fr/seaborn-cheats/)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset("tips")
tips
```

## Displot

Para distribuciones

```python
sns.displot(data=tips,
            x="total_bill",
            hue="sex",
            kind="kde",
            legend=True,
            palette="dark")
plt.show()
```

```python
sns.displot(data=tips,
            x="total_bill",
            y="tip",
            hue="sex" # Agrupamiento
            )
plt.show()
```

## Histplot

```python
sns.histplot(tips, x="tip", bins=15,
             cumulative=False, # Ir sumando los valores
             hue="sex", # Agrupamiento
             stat="count", # En vez del conteo muestra la estadistica (count, frequency, probability, percent, dencity)
             multiple="dodge", # Como se comparan los histogramas(layer, stack, dodge, fill)
             )
plt.show()
```

## Kdeplot (Densidad)

```python
sns.kdeplot(tips, x="tip", hue="sex",
            cumulative=False,
            fill=True, # Relleno
            bw_adjust=1, # Ajustar la grafica a los parametros de los ejes
            )
```

## Ecdfplot 

```python
sns.ecdfplot(tips, x="tip", hue="sex")
plt.show()
```

## Countplot

```python
sns.countplot(tips, x="day", hue="sex")
plt.show()
```

## Stripplot

```python
sns.stripplot(tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

## Swarmplot

```python
sns.swarmplot(tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

## Boxplot

```python
sns.boxplot(tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(tips, x="day", y="total_bill", hue="sex", dodge=True, palette="dark")
plt.show()
```

## Violinplot

```python
sns.violinplot(tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(tips, x="day", y="total_bill", hue="sex", dodge=True, palette="pastel")
plt.show()
```

## Catplot

```python
sns.catplot(tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
sns.catplot(tips, x="day", y="total_bill", hue="sex", dodge=True, col="time")
plt.show()
```

## ScatterPlot

```python
markers = {"Male":"D", "Female":"s"}
sns.scatterplot(tips, x="total_bill", y="tip", hue="day", style="sex",
                size="size",
                markers=markers)
plt.legend(loc="center", bbox_to_anchor=(1.12, 0.5))
plt.show()
```

## Lineplot

```python
sns.lineplot(tips, x="total_bill", y="tip", hue="day", style="sex")
plt.legend(loc="center", bbox_to_anchor=(1.12, 0.5))
plt.show()
```

## Relplot

```python
sns.relplot(tips, x="total_bill", y="tip", hue="day", style="sex", kind="scatter",
            col="time")
# plt.legend(loc="center", bbox_to_anchor=(1.12, 0.5))
plt.show()
```

## Jointplot

```python
sns.jointplot(tips, x="total_bill", y="tip", hue="sex", kind="scatter")
plt.show()
```

```python
sns.jointplot(tips, x="total_bill", y="tip", hue="sex", kind="hist")
plt.show()
```

```python
sns.jointplot(tips, x="total_bill", y="tip", hue="sex", kind="hist", 
              marginal_ticks=True, marginal_kws=dict(bins=25, multiple="dodge"))
plt.show()
```

## Pairplot

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm", corner=True, kind="reg")
plt.show()
```

## Regplot

```python
sns.regplot(tips, x="total_bill", y="tip")
```

## Heatmap

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), cbar=False)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap="coolwarm", linewidths=5,
            vmin=-1, vmax=1)
plt.show()
```
