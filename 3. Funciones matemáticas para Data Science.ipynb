## ¬øQu√© es una funci√≥n?

Es una regla en donde se obtiene un valor $y$ dado $x$, una funci√≥n solo es funci√≥n si para un $x$ solo hay una $y$ posible.

En otras palabras, la variable dependiente depende, valga la redundancia, de una o m√°s variables independientes

$y=f(x)$





Por ejemplo se puede obtener el tiempo que demora de un lugar a otro dada la distancia y la velocidad.


## Tipos de variables



- Cualitativas
    
    - Nominales: Son las que se les asigna una cualidad. Por ejemplo un color.

    - Ordinales: Que tienen un orden. Por ejemplo: alto, medio, bajo.

    - Binarias: Representan estados, que toman dos valores. Por ejemplo: Existe, g√©nero.

- Cuantitativas 

    - Discretas: Son finitas y toman ciertos valores, pueden verse como variables separadas por un "paso". Por ejemplo, una persona puede tener 1,  2 o 3 amigos, nuna 3.5.

    - Continuas: Pueden verse como infinitos dentro de los reales en un rango establecido. Por ejemplo, la estatura.

## Funciones algebraicas lineales

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Funci√≥n lineal

Tiene la forma:$$f(x)=mx+b$$

donde $m$ y $b \in R $

$m$ puede ser calculada por:

$$m=\frac{y_2-y_1}{x_2-x_1}$$

$m = \tan(\theta)$

Y $b$ es el punto de corte con el eje$y$.

$Dom_f=(-\inf,\inf)$

$Ran_f=(-\inf,\inf)$

```python
N = 100

m = np.random.uniform(-5, 10)
b = np.random.uniform(-10, 10)

def f(x):
    return m*x+b

x = np.linspace(-10, 10, num=N)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"$y={m}x+({b})$")
ax.grid()
plt.show()
```

### Funci√≥n polin√≥mica

$$P(x)=a_{n}x^{n}+a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x^{1}+a_1$$

Son de grado n

Ejemplo:

$$P(x)=2x^7-x^4+3x^2+4$$

```python
# Ejemplo de polinomio grado 7
def P(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

x = np.linspace(-10, 10, num=100)
y = P(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"$P(x)=2x^7-x^4+3x^2+4$")
ax.grid()
plt.show()
```

### Funciones seccionadas

Son funciones que est√°n dadas por intervalos. Por ejemplo:

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para } x < 0 \\
        1,  &\quad\text{para } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
    Y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            Y[idx]=1
    return Y

x = np.linspace(-10, 10, num=100) # Recordemos que los logaritmos no reciben n√∫meros negativos
y = H(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
plt.show()
```

## Funciones compuestas

Una funci√≥n compuesta se obtiene cuando se pasa como argumento de la funci√≥n principal el valor de otra funci√≥n. En otras palabras:

$$g(f(x))$$ 

```python
def f(x):
    return x**2

def g(x):
    return np.sin(x)

x = np.linspace(-10, 10, num=1000)
y = g(f(x))

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title(f"$y=\sin(x^2)$")
ax.grid()
plt.show()
```

## Manipulaci√≥n de funciones

### Desplazamiento

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x,y,label="Modificada")
ax.plot(x, f(x),label="Original")
ax.legend()
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f((1/c)*x)


fig, ax = plt.subplots()
ax.plot(x,y,label="Modificada")
ax.plot(x, f(x),label="Original")
ax.legend()
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(-x)

fig, ax = plt.subplots()
ax.plot(x,y,label="Modificada")
ax.plot(x, f(x),label="Original")
ax.legend()
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Regresi√≥n lineal simple

$$y=mx+b$$
$$y=B_0+B_1*X$$

Busca predecir a partir de una formula el valor de un


```python
from numpy.linalg import lstsq

x= np.array([1.2,2,3.2,2.5,5,6,4,8])
y= np.array([2,3,3.4,3.1,4,4.7,3.8,7])

m, b = lstsq(np.vstack([x, np.ones(len(x))]).T, y, rcond=None)[0]

print("Pendiente (m):", m)
print("Intercepto (b):", b)
```

```python
fig, ax = plt.subplots()
plt.scatter(x,y, color = 'mediumblue', alpha=0.6)
ax.set_xlabel('x',fontsize=16)
ax.set_ylabel('y',fontsize=16)
ax.axline((0, b), slope=m, color='red', linewidth=2)
plt.show()
```
